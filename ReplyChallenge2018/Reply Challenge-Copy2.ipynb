# Solution
* data -> contain all data about providers
* projects -> contain all data about projects

```python
def highestServiceNeed(pro):
    needs    = projects[pro,2:S+2]
    mostNeed = np.argmax(needs)
    #print("needs=",needs," mostNeed", mostNeed)
    if np.max(needs)<=0:
        return -1
    return mostNeed
    
def highestServiceOffer(need):
    services = data[0:data.shape[0],5+need]
    giveMore = np.argmax(services)
    #print("Services=",services," GiveMore", giveMore)
    if data[giveMore,3]==0:
        return -1
    return giveMore
    
def giveService(pro,serv):
    data[serv,3] -= 1 #decrement available packages
    for i in range(S):
        projects[pro,i+2] -= data[serv,i+5]*100
    checkDataAvailable()

def checkDataAvailable():
    for i in range(data.shape[0]):
        if data[i,3] < 100:
            data[i,5:5+S] = 0
            
def printRow(v):
    stringa = ""
    for i in range(len(v)):
        if v[i]!=0:
            stringa += str(int(data[i,0])) +" "+str(int(data[i,1]))+" "+str(int(v[i]))+" "
    stringa=stringa[0:len(stringa)-1]
    stringa=stringa+"\n"
    return stringa
```

```python
import codecs

file = codecs.open("third_adventure2.txt", "w","UTF-8") 

for pr in range(P):
    print(pr)
    check = True
    quantity = np.zeros([data.shape[0]])
    while check==True:
        mostNeed = highestServiceNeed(pr)
        if mostNeed >= 0:
            giveMore = highestServiceOffer(mostNeed)
            if giveMore >= 0:
                giveService(pr,giveMore)
                quantity[giveMore]+=100
            else:
                check=False
        else:
            check=False
    file.write(printRow(quantity))
            
file.close()
```
